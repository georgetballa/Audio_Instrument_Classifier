```python
import seaborn as sns
box = data[['instrument_family_str','pitch']]
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.titlecolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)

ax = sns.boxplot(x = 'instrument_family_str', y = 'pitch', data=box, palette='YlOrBr')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
for line in ax.get_lines():
    line.set_color('w')
    
for i,box in enumerate(ax.artists):
    box.set_edgecolor('w')
    
plt.title('Instrument Pitch Distribution')
plt.ylabel('Pitch')
plt.xlabel('Instrument Family')
plt.savefig(f'/home/george/Desktop/Instrument_Classifier/flask_app/instrument_classification/static/images/boxplot.png',pad_inches = 0,bbox_inches='tight', transparent=True, dpi=200)
            
```
